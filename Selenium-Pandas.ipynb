```python
# Selenium - Controlar o navegador - abrir fechar, tela cheia, etc
# criar navegador, entrar em sites, encontrar informações, fechar o navegador
# importando a biblioteca selenium
from selenium import webdriver # webdriver chrome - chromedriver
from selenium.webdriver.common.keys import Keys # importar as teclas
from selenium.webdriver.common.by import By # importar para o selenium achar elementos

navegador = webdriver.Chrome() # criando um navegador Google Chrome


# Passo 1 - Pegar a cotação do dólar

# selenium espera o navegador/site carregar
navegador.get("https://www.google.com.br/") # entrar no site
navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("cotação dólar") 
navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER) # apertar tecla enter
# find_element acha o elemento pelo caminho - inspecionar elemento
# o método send_keys escreve 
dolar = navegador.find_element(By.XPATH, '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value') # pegar o valor da cotação do dólar
print(dolar)



# Passo 2 - Pegar cotação do euro

navegador.get("https://www.google.com.br/") # entrar no site
navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("cotação euro") 
navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER) 
# find_element acha o elemento pelo caminho - inspecionar elemento
# o método send_keys escreve 
euro = navegador.find_element(By.XPATH, '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(euro)

# Passo 3 - Pegar cotação do ouro

navegador.get("https://www.melhorcambio.com/ouro-hoje") # entrar no site
ouro = navegador.find_element(By.XPATH, '//*[@id="comercial"]').get_attribute("value").replace(',', '.') # capturar o valor do ouro pelo campo 'value'
print(ouro)

navegador.quit()

```

```python
# Passo 4 - Importar e atualizar as cotações na base de dados
import pandas as pd

tabela = pd.read_excel("Produtos.xlsx") # ler o arquivo .xlsx
display(tabela)
```

- Atualizando os preços e o cálculo do Preço Final

```python
# Passo 5 - Calcular o preço com base na nova cotação

tabela.loc[tabela['Moeda'] == 'Dólar', 'Cotação'] = float(dolar) # método loc localiza para editar
tabela.loc[tabela['Moeda'] == 'Euro', 'Cotação'] = float(euro)
tabela.loc[tabela['Moeda'] == 'Ouro', 'Cotação'] = float(ouro)

# Cálculo do preço de compra
tabela["Preço de Compra"] = (tabela["Preço Original"] * tabela["Cotação"])
# Cálculo do preço de venda
tabela["Preço de Venda"] = (tabela["Preço de Compra"] * tabela["Margem"])

display(tabela)
```

```python
# Passo 6 - Salvar a base atualizada

tabela.to_excel("Produtos_Atualizados.xlsx", index = False) # exportar a tabela atualizada para um arquivo excel
# retira o índice para não aparecer eles no arquivo final
```
